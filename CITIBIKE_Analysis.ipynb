```python
import pandas as pd
import os as csv
```

```python
# Path to February data
path1 = "JC-202002-citibike-tripdata.csv/JC-202002-citibike-tripdata.csv"
# Read csv file
csv_file1 = pd.read_csv(path1)
# Turn csv file to dataframe
February = pd.DataFrame(csv_file1)
February
```

```python
# Path to March data
path2 = "JC-202003-citibike-tripdata.csv/JC-202003-citibike-tripdata.csv"
# Read csv file
csv_file2 = pd.read_csv(path2)
# Turn csv file to dataframe
March = pd.DataFrame(csv_file2)
March
```

```python
# Path to April data
path3 = "JC-202004-citibike-tripdata.csv/JC-202004-citibike-tripdata.csv"
# Read csv file
csv_file3 = pd.read_csv(path3)
# Turn csv file to dataframe
April = pd.DataFrame(csv_file3)
April
```

```python
# Append March dataframe to February dataframe
February_March = February.append(March, ignore_index=True)
February_March
```

```python
# Append April dataframe to February_March dataframe
February_April = February_March.append(April, ignore_index=True)
February_April
```

```python
# Create new column with date by splitting startime column and using first string
February_April["Date"] = February_April["starttime"].str.split(' ', 1, expand=True)[0]
February_April
```

```python
# Save dataframe as csv file without index
February_April.to_csv("Feb_Apr1.csv", index=False)
```
